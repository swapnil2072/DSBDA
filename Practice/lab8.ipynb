```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
dataset = sns.load_dataset("Titanic")
```

```python
dataset
```

```python
dataset['age'] = dataset['age'].fillna(dataset['age'].mean())
dataset['embarked'] = dataset['embarked'].fillna(dataset['embarked'].mode()[0])
dataset['deck'] = dataset['deck'].fillna(dataset['deck'].mode()[0])
dataset['embark_town'] = dataset['embark_town'].fillna(dataset['embark_town'].mode()[0])
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(dataset)
```

```python
sns.distplot(dataset['fare'])
plt.show()
```

```python
sns.pairplot(dataset,hue='sex')
```

```python
sns.catplot(x='sex',hue='survived',kind='count',data=dataset)
```

```python
sns.boxplot(x='sex',y='age',hue='survived',data = dataset)
```

```python
sns.histplot(dataset['fare'])
plt.show()
```

```python
sns.displot(data=dataset,kde=True)
```

```python
sns.violinplot(x ="sex", y ="age", hue ="survived",data = dataset,)
```
